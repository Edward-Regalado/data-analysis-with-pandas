```python
station = pd.read_csv('./station.csv')
trips = pd.read_csv('./trip.csv', error_bad_lines=False)
weather = pd.read_csv('./weather.csv')
```

```python
station
```

```python
## 1. Average Trip Duration for borrowed bicycle
avg_trip_dur = str(round(trips['tripduration'].mean()/60, 2)).split('.')
avg_trip_dur[1] = str(round((60*int(avg_trip_dur[1])/100)))
f"The average trip duration for a borrowed bike is {avg_trip_dur[0]}:{avg_trip_dur[1]}!"
```

```python
## 2. Most common age of a bicyle sharer
most_common_bike_year = 2021 - int(trips['birthyear'].mode())
f'The most common age of of a bicycle sharer is {most_common_bike_year}.'
```

```python
## 3. Find the average precipitation per month, and the median precipitation.
average_precip = weather[['Date', 'Precipitation_In']]
precip_data = []
for i in range(12):
  month_data = [row.Precipitation_In for row in average_precip.iloc if int(row.Date.split('/')[0]) == i+1]
  month_data = np.array(month_data)
  precip_data.append(f'{calendar.month_abbr[i+1]} - mean precipitation: {round(np.mean(month_data), 2)} inches / median precipitation: {np.median(month_data)} inches')

precip_data
```

```python
## 4. What's the average number of bike at a given bike station?
average_bike = round(station['current_dockcount'].mean())
f'The average number of Bikes at a bike station is {average_bike}.'
```

```python
## 5. When a bike station is modified, is it more likely that it'll lose bikes or gain bikes? How do you know?
mod_dockcount = round(station[['name', 'install_dockcount', 'current_dockcount', 'modification_date']].sort_values('modification_date').mean())

# print iloc(integer location from index 0-2)
a = mod_dockcount.iloc[0:2]
column = pd.DataFrame(a, columns=['Modified Dockcounts gain bikes on average'])
column
```
